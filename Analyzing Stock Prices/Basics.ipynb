<h1>Working with dictionaries and pandas data frames</h1>

```python
from csv import reader
import pandas as pd
import os

stock_prices = {}
for fn in os.listdir("prices"):
    name = fn.split(".")[0]
    stock_prices[name] = pd.read_csv(os.path.join("prices", fn))

display(stock_prices["amzn"])
```

```python
cpDict = {}
for name in stock_prices:
        cpDict[name] = stock_prices[name]["close"].mean()
print(cpDict["blfs"])
```

```python
print("min: " + str(min(cpDict, key=cpDict.get)) + ": " + str(cpDict[str(min(cpDict, key=cpDict.get))]))
print("max: " + str(max(cpDict, key=cpDict.get)) + ": " + str(cpDict[str(max(cpDict, key=cpDict.get))]))
```

```python
from collections import defaultdict
dateVolDict = defaultdict(list)
for name in stock_prices:
    for index, row in stock_prices[name].iterrows():
        dateVolDict[row["date"]].append([row["volume"], name])
```

```python
from operator import itemgetter
mostTradedByDayDict = {}
for date in dateVolDict:
    mostTradedByDayDict[date] = max(dateVolDict[date],key=itemgetter(0))[1]
print(mostTradedByDayDict)
```

```python
tradeVolumeByDayDict = {}
for date in dateVolDict:
    tradeVolumeByDayDict[date] = sum(i[0] for i in dateVolDict[date])
print(sorted(tradeVolumeByDayDict, key=tradeVolumeByDayDict.get, reverse=True)[:10])
```

<h1>Using dictionaries we can determine that the most profitable stocks in the list are the following</h1>

```python
stockProfitabilityDict = {}
for name in stock_prices:
    stockProfitabilityDict[name] = (stock_prices[name].iloc[-1]["close"] - stock_prices[name].iloc[0]["close"])/stock_prices[name].iloc[0]["close"]
print(sorted(stockProfitabilityDict, key=stockProfitabilityDict.get, reverse=True)[:10])
```
