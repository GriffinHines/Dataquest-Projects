```python
from csv import reader
def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line after each row

    if rows_and_columns:
        print('Number of rows:', len(dataset))
       # print('Number of columns:', len(dataset[0]))
```

<h3>The data:</h3>

```python
with open('AppleStore.csv', encoding='utf8') as read_obj:
    # pass the file object to reader() to get the reader object
    csv_reader = reader(read_obj)
    # Pass reader object to list() to get a list of lists
    appleStore = list(csv_reader)
with open('googleplaystore.csv', encoding='utf8') as read_obj:
    # pass the file object to reader() to get the reader object
    csv_reader = reader(read_obj)
    # Pass reader object to list() to get a list of lists
    googlePlayStore = list(csv_reader)

explore_data(appleStore, 0, 3, True)
print("----------------------------------")
explore_data(googlePlayStore, 0, 3, True)
```

<h3>Apple store column names are not all entirely clear, below is a description of each:</h3>


    "id" : App ID

    "track_name": App Name

    "size_bytes": Size (in Bytes)

    "currency": Currency Type

    "price": Price amount

    "ratingcounttot": User Rating counts (for all version)

    "ratingcountver": User Rating counts (for current version)

    "user_rating" : Average User Rating value (for all version)

    "userratingver": Average User Rating value (for current version)

    "ver" : Latest version code

    "cont_rating": Content Rating

    "prime_genre": Primary Genre

    "sup_devices.num": Number of supporting devices

    "ipadSc_urls.num": Number of screenshots showed for display

    "lang.num": Number of supported languages

    "vpp_lic": Vpp Device Based Licensing Enabled

```python
#google play store has some duplicate entries
for app in googlePlayStore:
    name = app[0]
    if name == 'Instagram':
        print(app)
```

```python
#number of duplicates. Highest number of reviews for each duplicate is calculated to keep only the duplicate with the highest number of reviews (this is likely the most recent data)
names = set()
duplicateReviewsMax = {}
numberOfDuplicates = 0
for app in googlePlayStore[1:]:
    name = app[0]
    if name in names:
        numberOfDuplicates += 1
        if name not in duplicateReviewsMax or duplicateReviewsMax[name] < app[3]:
            duplicateReviewsMax[name] = app[3]
    else:
        names.add(name)
        duplicateReviewsMax[name] = app[3]
```

```python
android_clean = []
already_added = set()
for app in googlePlayStore[1:]:
    name = app[0]
    n_reviews = float(app[3])
    if(float(duplicateReviewsMax[name]) == n_reviews and name not in already_added):
        android_clean.append(app)
        already_added.add(name)
explore_data(android_clean, 0, 3, True)
```

```python
def checkIfStringIsEnglish(inputString):
    count = 0
    for character in inputString:
        if ord(character) > 127:
            count+=1
    if count > 3:
        return False
    return True

checkIfStringIsEnglish('Instachat ðŸ˜œ')
```

<h3>Clean non english apps from data</h3>

```python
android_clean_english = []
for app in android_clean:
    if(checkIfStringIsEnglish(app[0])):
        android_clean_english.append(app)
        
apple_english = []
for app in appleStore[1:]:
    if(checkIfStringIsEnglish(app[0])):
        apple_english.append(app)

explore_data(apple_english, 0, 3, True)
explore_data(android_clean_english, 0, 3, True)
```

<h3>Free apps</h3>

```python
free_apple_apps = []
for app in apple_english:
    if app[4] == "0.0":
        free_apple_apps.append(app)
free_android_apps = []
for app in android_clean_english:
    if app[7] == "0":
        free_android_apps.append(app)
        
explore_data(free_apple_apps, 0, 3, True)
explore_data(free_android_apps, 0, 3, True)
```

<h3>Frequency tables based on genre/category<h3>

```python
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
        
def freq_table(dataset, index):
    table = {}
    for app in dataset:
        if(app[index] not in table):
            table[app[index]] = 1
        else:
            table[app[index]] += 1
    return table

display_table(apple_english, 11)
print("------------------------------------------------")
display_table(free_android_apps, 9)
print("------------------------------------------------")
display_table(free_android_apps, 1)
```

```python
prime_genreFreqTable = freq_table(apple_english, 11)
genreToRatings = {}
for genre in prime_genreFreqTable:
    total = 0
    len_genre = 0
    for app in apple_english:
        genre_app = app[11]
        if(genre_app == genre):
            num = ""
            for c in app[5]:
                if c.isdigit():
                    num = num + c
            total+=float(num)
            len_genre+=1
    avg = int(total/len_genre)
    print(genre + " average rating count = " + str(avg))

table_display = []
for key in genreToRatings:
    key_val_as_tuple = (genreToRatings[key], key)
    table_display.append(key_val_as_tuple)

table_sorted = sorted(table_display, reverse = True)
for entry in table_sorted:
    print(entry[1], ':', entry[0])
```
