```python
%%capture
%load_ext sql
%sql sqlite:///factbook.db
```

```python
%%sql
SELECT *
  FROM sqlite_master
 WHERE type='table';
```

```python
%%sql
SELECT *
    FROM facts
 LIMIT 5;
```

```python
%%sql
SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
    FROM facts;
```

```python
%%sql
SELECT name
    FROM facts
 WHERE population == 0;
```

```python
%%sql
SELECT name
    FROM facts
 WHERE population == (SELECT MAX(population) FROM facts);
```

```python
%%sql
SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
    FROM facts
 WHERE name != 'World' AND name != 'Antarctica';
```

```python
%%sql
SELECT AVG(population), AVG(area)
    FROM facts
 WHERE name != 'World' AND name != 'Antarctica';
```

```python
%%sql
SELECT name as 'Densely Populated Countries', population, area, CAST(population AS FLOAT)/CAST(area AS FLOAT) as 'population/area'
    FROM facts
 WHERE population > (SELECT AVG(population) FROM facts WHERE name != 'World' AND name != 'Antarctica')
    AND area < (SELECT AVG(area) FROM facts WHERE name != 'World' AND name != 'Antarctica')
ORDER BY CAST(population AS FLOAT)/CAST(area AS FLOAT) DESC;
```

```python
%%sql
SELECT name, CAST(area_water AS FLOAT)/CAST(area_land AS FLOAT) AS "Ratio of Water to Land"
    FROM facts
 ORDER BY CAST(area_water AS FLOAT)/CAST(area_land AS FLOAT)
```
