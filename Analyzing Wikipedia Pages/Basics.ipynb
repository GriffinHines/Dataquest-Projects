<h1>Custom functions to grep wikipedia articles for the occurances of strings, utilizing multithreading</h1>

```python
import os
from csv import reader
file_names = os.listdir("wiki")
count = 0
for i in file_names:
    count +=1
    print(i)
print(count)
    
with open(os.path.join("wiki", file_names[0]), encoding='utf8') as read_obj:
    # pass the file object to reader() to get the reader object
    csv_reader = reader(read_obj)
    # Pass reader object to list() to get a list of lists
    BayOfCon = list(csv_reader)
for row in BayOfCon:
    print(row)
```

```python
import math
import functools
from multiprocessing import Pool

def make_chunks(data, num_chunks):
    chunk_size = math.ceil(len(data) / num_chunks)
    return [data[i:i+chunk_size] for i in range(0, len(data), chunk_size)]

def map_reduce(data, num_processes, mapper, reducer):
    chunks = make_chunks(data, num_processes)
    pool = Pool(num_processes)
    chunk_results = pool.map(mapper, chunks)
    return functools.reduce(reducer, chunk_results)

def count_lines(file_name):
    summation = 0
    for file in file_name:
        summation += sum(1 for line in open(os.path.join("wiki", file)))
    return summation
                                  
def reducer(a, b):
    return a + b

lines = map_reduce(file_names, 9, count_lines, reducer)
print(lines)
```

```python
def lineIndexDict(string):
    indexDict = {}
    for file in file_names:
        lineNum = 0
        lineNumList = []
        found = False
        for line in open(os.path.join("wiki", file)):
            lineNum += 1
            if string in line:
                found = True
                lineNumList.append(lineNum)
        if(found):
            indexDict[file] = lineNumList
    return indexDict

def indexDictReducer(dict1, dict2):
    for val in dict2:
        if dict1[val]:
            dict1[val] = dict1[val] + dict2[val]
        else:
            dict1[val] = dict2[val]
    return dict1

lines = map_reduce("data", 3, lineIndexDict, indexDictReducer)
print(lines)
```

```python
def lineIndexDictCaseInsensitive(string):
    indexDict = {}
    for file in file_names:
        lineNum = 0
        lineNumList = []
        found = False
        for line in open(os.path.join("wiki", file)):
            lineNum += 1
            if string in line.lower():
                found = True
                lineNumList.append(lineNum)
        if(found):
            indexDict[file] = lineNumList
    return indexDict

linesCaseInsensitive = map_reduce("data", 3, lineIndexDictCaseInsensitive, indexDictReducer)

for i in lines:
    if len(lines.get(i)) < len(linesCaseInsensitive.get(i)):
        print(str(i) + " " + str(linesCaseInsensitive[i]))
```

```python
def lineIndexDictCaseInsensitiveSpecific(string):
    indexDict = {}
    for file in file_names:
        lineNum = 0
        lineNumList = []
        found = False
        for line in open(os.path.join("wiki", file)):
            lineNum += 1
            if string in line.lower():
                found = True
                index = line.lower().find(string)
                context = getContext(line, index)
                while(index != -1):
                    lineNumList.append((lineNum, index, context))
                    index = line.lower().find(string, index+1)
                    context = getContext(line, index)
        if(found):
            indexDict[file] = lineNumList
    return indexDict

#get context of index
def getContext(line, index):
    if(index < 20):
        start = 0
    else:
        start = index - 20
    if(index+20 > len(line)):
        end = len(line)
    else:
        end = index+20
    return line[int(start):int(end)]
linesCaseInsensitiveSpecific = map_reduce("data", 1, lineIndexDictCaseInsensitiveSpecific, indexDictReducer)
print(linesCaseInsensitiveSpecific)
```
